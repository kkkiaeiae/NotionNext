```python
import os
import glob
from pydub import AudioSegment

# 1. 定位 Gradio 临时文件夹下的所有 audio.wav 文件
file_pattern = '/tmp/gradio/*/audio.wav'
audio_files = glob.glob(file_pattern)

# 2. 核心步骤：严格按照文件的生成时间排序（确保拼出来的音频顺序是对的）
audio_files.sort(key=os.path.getmtime)

print(f"一共找到了 {len(audio_files)} 个音频片段，正在努力拼接中，请稍候...")

# 3. 开始无缝拼接
combined_audio = AudioSegment.empty()
for file in audio_files:
    combined_audio += AudioSegment.from_wav(file)

# 4. 导出为一整个完整的音频文件，直接放在外面方便下载
output_path = '/content/完整版21分钟干音.wav'
combined_audio.export(output_path, format="wav")

print(f"🎉 拼接大功告成！文件已保存为：{output_path}")
```
